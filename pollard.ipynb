```python
def primorialize(B):
    primorials = [];
    counter = 0;
    i = 2;
    while counter < B:
        if is_prime(i):
            primorials.append(i)
            counter+=1
        i+=1
    return primorials
```

```python
def primorial(B):
    primorials = [];
    result = 1
    counter = 0;
    i = 2;
    while counter < B:
        if is_prime(i):
            primorials.append(i)
            counter+=1
        i+=1
    for x in primorials:
        result = result * x
    return result
```

```python
def factors(n):
    factors = []
    for i in range(1, n):
        if n%i==0:
            if is_square(i):
                factors.append(i)
            if is_prime(i):
                factors.append(i)
    return factors
```

```python
#copied from internet remember to modify
import math
def is_square(integer):
    root = math.sqrt(integer)
    return integer == int(root + 0.5) ** 2
```

```python
#copied from internet remember to modify
import math
def is_prime(n):
    if n % 2 == 0 and n > 2: 
        return False
    for i in range(3, int(math.sqrt(n)) + 1, 2):
        if n % i == 0:
            return False
    return True
```
