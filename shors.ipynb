```python
def find_t(N,g):
    powers = [g]
    #print(powers)
    count = 1
    t=0
    temp = 0
    while(True):
        temp = (pow(g,count+1)) % N
        print(str(g)+"^"+str(count+1)+" mod "+str(N)+" = "+str(temp))
        print(temp)
        powers.append(temp)
        #print(powers)
        t+=1
        count+=1
        if(powers[0] == temp):
            print(str(temp) + " = " + str(powers[0]))
            break
        #print(str(t))
    return [t,powers]
    
```

```python
import math
def shors(N,g):
    t=3
    
    while(t%2 != 0):
        results = find_t(N,g)
        t = results[0]
        powers = results[1]
        print(powers)
        #print(str(t))
    a = powers[int((t/2)-1)] - 1
    b = powers[int((t/2)-1)] + 1
    
    p = math.gcd(a,N)
    print("GCD("+str(a)+", "+str(N)+") = " +str(p) )
    
    if p == 1:
        print("Try with b...")
        q = math.gcd(b,N)
        print("GCD("+str(b)+", "+str(N)+") = " +str(q) )

        p = int(N/q)
        
    else:
        q = int(N/p)
    if p == 1 or q == 1:
            print(str(g) + " can't be used to factor " + str(N) + " with shors")
    
    
    print( "t = " + str(t))
    print( "a = " + str(a))
    print( "b = " + str(b))
    print( "p = " + str(p))
    print( "q = " + str(q))
    
        
    
```
